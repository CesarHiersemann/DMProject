```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import keras
import pandas as pd
from keras.preprocessing import sequence
```

Load datasets

```python
train = np.asarray(pd.read_csv('./data/train.csv'))
test = np.asarray(pd.read_csv('./data/test.csv'))
```

Concatenate train and test

```python
data = np.vstack((train[:,:-1],test))
```

```python
data.shape
```

Start by putting the continuous input features in X and labelled output from training set in y

```python
X = data[:,-14:].astype(np.float32)
y = train[:,-1].astype(np.float32)
```

One hot encoding of all the 116 cathegorical variables

```python
for i in range(1,117):
    fac = pd.factorize(data[:,i])
    X = np.hstack((keras.utils.np_utils.to_categorical(fac[0]),X))
```

Split train and test again

```python
X_train = X[:train.shape[0]]
X_test = X[train.shape[0]:]
```

```python
X_train.shape
```

Final input has 1190 features

```python
#1180 1176
model = keras.models.Sequential()
model.add(keras.layers.Dense(10,activation="relu",input_dim=1153))
#model.add(keras.layers.Dense(200,activation="relu"))
#model.add(keras.layers.Dense(10,activation="relu"))
#model.add(keras.layers.Dense(10,activation="relu"))
model.add(keras.layers.Dense(1))
model.compile(optimizer='adadelta',loss='mae')
```

```python
#normalize cont
X_train[:,-14:] = (X_train[:,-14:] - np.mean(X_train[:,-14:],axis=0))/np.std(X_train[:,-14:],axis=0)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(400,activation="relu",input_dim=1190))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(200,activation="relu"))
model.add(keras.layers.Dropout(0.2))
#model.add(keras.layers.Dense(10,activation="relu"))
#model.add(keras.layers.Dense(10,activation="relu"))
model.add(keras.layers.Dense(1))
model.compile(optimizer='rmsprop',loss='mae')
```

```python
model.fit(X_train,y,nb_epoch=100,batch_size=32,validation_split=0.2)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:,0]
```

```python
df = pd.DataFrame({'id': test[:,0], 'loss': y_pred[:,0]})
df.to_csv('submission.csv', index = False)
```
